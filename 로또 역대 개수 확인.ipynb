```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
from tqdm import tqdm,trange
from time import sleep
```

```python
def crawling_sites(url):
    response = requests.request('Get', url)
    soup = bs(response.content, 'html.parser')
    
    titles = soup.select('div.winning_number')
    
    num_title = []
    
    for one in titles:
        num_title.append(one.get_text())
    
    return num_title
```

```python

ball_list = []

with tqdm(total =100) as pbar:
    for train in range(944,1032):
    
        html ='https://search.naver.com/search.naver?where=nexearch&sm=top_sly.hst&fbm=0&acr=1&acq=%EB%A1%9C&qdt=0&ie=utf8&query='+str(train)+'%ED%9A%8C+%EB%A1%9C%EB%98%90%EB%8B%B9%EC%B2%A8%EB%B2%88%ED%98%B8'
    
        ball_list.append(crawling_sites(html))
    
        train+=1

        sleep(0.1)
        pbar.update(1)
```

```python
for i, a in enumerate(ball_list):
    print(i,a)
```
